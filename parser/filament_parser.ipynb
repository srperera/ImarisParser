```python
from utils.data import load_yaml
from scripts_v2.process_filaments import run_filament_parser_parallel
```

```python
config_path = "config/config_filaments.yaml"
config = load_yaml(config_path)
```

```python
if config["extract_objects"]:
    if not isinstance(config["valid_filaments"], list):
        print(f"[info] -- running all available surfaces")

        if int(config["valid_filaments"]) == -1:
            run_filament_parser_parallel(
                data_dirs=config["data_dir"],
                save_dirs=config["save_dir"],
                cpu_cores=config["cpu_cores"],
                filament_ids=None,
            )

    elif isinstance(config["valid_filaments"], list):
        print(f"[info] -- running specified surfaces")

        valid_filaments = [int(x) for x in config["valid_filaments"]]

        run_filament_parser_parallel(
            data_dirs=config["data_dir"],
            save_dirs=config["save_dir"],
            cpu_cores=config["cpu_cores"],
            filament_ids=valid_filaments,
        )
    else:
        raise NotImplementedError
```
