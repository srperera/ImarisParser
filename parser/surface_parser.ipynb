```python
from utils.data import load_yaml
from scripts.process_surface import (
    run_surface_parser_parallel,
    run_surface_track_parser_parallel,
    run_surface_track_object_parser_parallel,
    run_surface_timestep_parser_parallel,
)
```

### Load Config

```python
config_path = "config/config_surface.yaml"
config = load_yaml(config_path)
```

### Extract Surface Object Information

```python
if config["extract_objects"]:
    if not isinstance(config["valid_surfaces"], list):
        print(f"[info] -- running all available surfaces")

        if int(config["valid_surfaces"]) == -1:
            run_surface_parser_parallel(
                data_dirs=config["data_dir"],
                save_dirs=config["save_dir"],
                cpu_cores=config["cpu_cores"],
                surface_ids=None,
            )

    elif isinstance(config["valid_surfaces"], list):
        print(f"[info] -- running specified surfaces")

        valid_surfaces = [int(x) for x in config["valid_surfaces"]]

        run_surface_parser_parallel(
            data_dirs=config["data_dir"],
            save_dirs=config["save_dir"],
            cpu_cores=config["cpu_cores"],
            surface_ids=valid_surfaces,
        )
    else:
        raise NotImplementedError
```

### Extract Surface Track Information

```python
if config["extract_tracks"]:
    if not isinstance(config["valid_surfaces"], list):
        print(f"[info] -- running all available surfaces")

        if int(config["valid_surfaces"]) == -1:
            run_surface_track_parser_parallel(
                data_dirs=config["data_dir"],
                save_dirs=config["save_dir"],
                cpu_cores=config["cpu_cores"],
                surface_ids=None,
            )

    elif isinstance(config["valid_surfaces"], list):
        print(f"[info] -- running specified surfaces")

        valid_surfaces = [int(x) for x in config["valid_surfaces"]]

        run_surface_track_parser_parallel(
            data_dirs=config["data_dir"],
            save_dirs=config["save_dir"],
            cpu_cores=config["cpu_cores"],
            surface_ids=valid_surfaces,
        )
    else:
        raise NotImplementedError
```

### Extract Surface Track Object Information

```python
if config["extract_surface_track_objects"]:
    if not isinstance(config["valid_surfaces"], list):
        print(f"[info] -- running all available surfaces")

        if int(config["valid_surfaces"]) == -1:
            run_surface_track_object_parser_parallel(
                data_dirs=config["data_dir"],
                save_dirs=config["save_dir"],
                cpu_cores=config["cpu_cores"],
                surface_ids=None,
            )

    elif isinstance(config["valid_surfaces"], list):
        print(f"[info] -- running specified surfaces")

        valid_surfaces = [int(x) for x in config["valid_surfaces"]]

        run_surface_track_object_parser_parallel(
            data_dirs=config["data_dir"],
            save_dirs=config["save_dir"],
            cpu_cores=config["cpu_cores"],
            surface_ids=valid_surfaces,
        )
    else:
        raise NotImplementedError
```

```python
if config["extract_specific_timestep"]:
    if not isinstance(config["valid_surfaces"], list):
        print(f"[info] -- running all available surfaces")

        if int(config["valid_surfaces"]) == -1:
            run_surface_timestep_parser_parallel(
                data_dirs=config["data_dir"],
                save_dirs=config["save_dir"],
                cpu_cores=config["cpu_cores"],
                surface_ids=None,
                time_step=config["time_index"],
            )

    elif isinstance(config["valid_surfaces"], list):
        print(f"[info] -- running specified surfaces")

        valid_surfaces = [int(x) for x in config["valid_surfaces"]]

        run_surface_timestep_parser_parallel(
            data_dirs=config["data_dir"],
            save_dirs=config["save_dir"],
            cpu_cores=config["cpu_cores"],
            surface_ids=valid_surfaces,
            time_step=config["time_index"],
        )
    else:
        raise NotImplementedError
```
