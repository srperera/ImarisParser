```python
from utils.data import load_yaml
from scripts.process_spots import (
    run_spot_track_object_parser_parallel,
    run_spot_track_parser_parallel,
)
```

### Load Config

```python
config_path = "config/config_spots.yaml"
config = load_yaml(config_path)
```

### Extract Spot Track Object Information

```python
if config["extract_track_objects"]:
    if not isinstance(config["valid_spots"], list):
        print(f"[info] -- running all available surfaces")

        if int(config["valid_spots"]) == -1:
            run_spot_track_object_parser_parallel(
                data_dirs=config["data_dir"],
                save_dirs=config["save_dir"],
                cpu_cores=config["cpu_cores"],
                spot_ids=None,
            )

    elif isinstance(config["valid_surfaces"], list):
        print(f"[info] -- running specified surfaces")

        valid_spots = [int(x) for x in config["valid_spots"]]

        run_spot_track_object_parser_parallel(
            data_dirs=config["data_dir"],
            save_dirs=config["save_dir"],
            cpu_cores=config["cpu_cores"],
            spot_ids=valid_spots,
        )
    else:
        raise NotImplementedError
```

### Extract Spot Track Information

```python
if config["extract_tracks"]:
    if not isinstance(config["valid_spots"], list):
        print(f"[info] -- running all available surfaces")

        if int(config["valid_spots"]) == -1:
            run_spot_track_parser_parallel(
                data_dirs=config["data_dir"],
                save_dirs=config["save_dir"],
                cpu_cores=config["cpu_cores"],
                spot_ids=None,
            )

    elif isinstance(config["valid_surfaces"], list):
        print(f"[info] -- running specified surfaces")

        valid_spots = [int(x) for x in config["valid_spots"]]

        run_spot_track_parser_parallel(
            data_dirs=config["data_dir"],
            save_dirs=config["save_dir"],
            cpu_cores=config["cpu_cores"],
            spot_ids=valid_spots,
        )
    else:
        raise NotImplementedError
```
